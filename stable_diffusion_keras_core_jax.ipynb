```python
import os

os.environ["KERAS_BACKEND"] = "jax"
```

```python
import keras_cv
import matplotlib.pyplot as plt
```

```python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
prompt = "photograph of an astronaut riding a horse, he is a space cowboy"
```

```python
# First run
%time images = model.text_to_image(prompt, batch_size=3)
```

```python
# Second run
%time images = model.text_to_image(prompt, batch_size=3)
```

```python
def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```
